## 2. Visualise data

##### Visualize numerical columns through the appropriate plots.

##### Age, Gender
##### StudyTimeWeekly, GPA

```python
# Bar plot of Age
plt.figure(figsize=(8, 6))
df['Age'].value_counts().sort_index().plot(kind='bar') # Plot the graph

# Add title and labels
plt.title('Bar Graph of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

##### Looking at the Bar Graph of ages it can be seen that the distribution of ages is roughly equally distributed

```python
# Pie chart of Gender distribution
plt.figure(figsize=(8, 8))
labels = ['Female', 'Male'] # Labels for pie graph
df['Gender'].value_counts().plot(kind='pie',labels = labels, autopct='%1.1f%%', startangle=140) # Plot pie graph

# Title and axis of graph
plt.title('Pie Chart of Gender Distribution')
plt.ylabel('')  # Hide the y-label to prevent any misinterpretting of graph
plt.show()
```

##### Interpretting the pie chart we can see that this dataset has roughly the same number of males and females

```python
# Box plot of StudyTimeWeekly
plt.figure(figsize=(8, 6))
sns.boxplot(y='StudyTimeWeekly', data=df) # Plot the box and whisker using seaborn
# Labelling the graph
plt.title('Box and Whisker Plot of Weekly Study Time of Students')
plt.ylabel('Study Time (Hours)')
plt.show()
```

##### Looking at the Box and Whisker plot we can see that the weekly study time is evenly distributed

```python
# Histogram of GPA
plt.figure(figsize=(8, 6))
plt.hist(df['GPA'], bins=10, edgecolor='black')
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

##### The distribution of GPA follows a normal distribution. So there does not seem to be any outliers for this feature.

```python
# Bar plot of Absences
plt.figure(figsize=(8, 6))
df['Absences'].value_counts().sort_index().plot(kind='bar') # Plot the graph

# Add title and labels
plt.title('Bar Graph of Absences')
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

## 4. Identify correlated variables

```python
# Calculate the correlation matrix
corr_matrix = df_drop_feat.corr()

# Create a mask using numpy's triu function
# This will create a mask that will cover the top right of the heatmap making the heatmap more readable
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Create a masked heatmap
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(corr_matrix, cmap = 'coolwarm', vmin = -1, vmax = 1, center = 0, annot=True, fmt=".2f", square=True, linewidths=.5, mask = mask)
plt.show()
```

##### Looking at the correlation matrix there is no strong correlation between the features at the moment. Where -1 means there is negative correlation, 0 meaning no correlation and 1 indiciating that there is a positive correlation. So when later when choosing models and performing more exploratory data analysis we could find out what other features to exclude from the model.
